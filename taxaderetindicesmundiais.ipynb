```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['^GSPC', '^IXIC', '^GDAXI', 'BOVA11.SA'] 
ind_data = pd.DataFrame()
for t in tickers:
    ind_data[t]= wb.DataReader(t, data_source='yahoo', start= '2009-1-1')['Adj Close']
```

```python
ind_data.head()
```

```python
ind_data.tail() 
```

```python
(ind_data / ind_data.iloc[0] * 100).plot(figsize = (15,6));
```

```python
ind_returns = (ind_data / ind_data.shift(1)) - 1
ind_returns
```

```python
anual=ind_returns.mean()*250
anual
```
